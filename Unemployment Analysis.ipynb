ANALYSIS

```python
plt.title = ("Indian Unemployment")
sns.histplot(x=" Estimated Employed", hue="Region.1", data=df)
plt.show()
```

```python
fig = px.histogram(df, x="Region", y=" Estimated Unemployment Rate (%)", title="Indian Unemployment Rate(state-wise)", template="plotly", color= 'Region')
fig.update_layout(xaxis={'categoryorder': 'total descending'})
fig.show()
```

```python
df.columns
```

```python
plt.Figure(figsize=(10,10))
plt.title="Indian Unemploment Rate"
sns.histplot(x=" Estimated Unemployment Rate (%)", hue="Region.1", data=df)
plt.show()
```

```python
unemployment= df[["Region","Region.1"," Estimated Unemployment Rate (%)"]]
figure = px.sunburst(unemployment, path= ["Region.1", "Region"],
                     values=" Estimated Unemployment Rate (%)",
                     width=600,height=600, color_continuous_scale="RdY1Gn",
                     title= "Unemployment Rate In INDIA")
figure.show()
```
